```python
import pandas as pd
df = pd.read_csv("./all_data_v1.csv")
df
```

```python
df['category'].value_counts()
```

```python
df['main_category'].value_counts()
```

```python
df['sub_category'].value_counts()
```

```python
category_mapping = {
    "women's clothing": "apparel_womens",
    "men's shoes": "footwear_mens",
    "appliances": "home_appliances",
    "kids' fashion": "apparel_kids",
    "tv, audio & cameras": "electronics_audio_video",
    "home & kitchen": "home_kitchen"
}

df['main_category'] = df['main_category'].replace(category_mapping)
df['main_category'].value_counts()
```

```python
df.drop(columns=['sub_category','image','link','actual_price','category'],inplace=True)
df
```

```python
df.info()
```

```python
df['discount_price'] = df['discount_price'].str.replace('₹', '').str.replace(',','').astype('float')
```

```python
df['no_of_ratings'] = df['no_of_ratings'].str.replace(',','').astype('int')
```

```python
df.info()
```

```python
df
```

```python
df['text_for_embedding'] =  df['main_category']+":"+df['name']
```

```python
df
```

```python
df.to_csv('./data_for_embedding.csv',index=False)
```
